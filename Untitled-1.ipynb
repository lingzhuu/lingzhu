```python
def quality_control(
    adata,
    mode="percentile",          # "percentile" or "absolute"
    groupby=None,               # per-sample thresholds (e.g., "sample" or "batch")
    # percentile mode (0–100)
    low_umi_pct=1,
    high_umi_pct=99,
    mt_cap_pct=20,
    # absolute mode
    min_umi=None,
    max_umi=None,
    # extras
    drop_mt_genes=True,):
    """
    Filter Visium/space spots by UMI, genes, and mt% using either percentiles or absolute thresholds.
    """
    #logger.info("Performing quality control...")

    obs = adata.obs

    adata.var['mt'] = adata.var_names.str.startswith(('MT-','mt-'))
    sc.pp.calculate_qc_metrics(adata, layer="count", qc_vars=['mt'], inplace=True)

    def pct_series(name, p):
        if p is None:
            return None
        if groupby is None:
            v = np.percentile(obs[name].to_numpy(), p)
            return pd.Series(v, index=obs.index)
        return obs.groupby(groupby)[name].transform(lambda x: np.percentile(x.to_numpy(), p))

    if mode == "percentile":
        low_umi_s   = pct_series("total_counts",   low_umi_pct)
        high_umi_s  = pct_series("total_counts",   high_umi_pct)

    elif mode == "absolute":
        def abs_series(val): return None if val is None else pd.Series(val, index=obs.index)
        low_umi_s   = abs_series(min_umi)
        high_umi_s  = abs_series(max_umi)
    else:
        raise ValueError("mode must be 'percentile' or 'absolute'")

    keep = pd.Series(True, index=obs.index)
    if low_umi_s   is not None: keep &= obs["total_counts"]  >= low_umi_s
    if high_umi_s  is not None: keep &= obs["total_counts"]  <= high_umi_s
    if mt_cap_pct    is not None: keep &= obs["pct_counts_mt"] <= mt_cap_pct

    adata_qc = adata[keep.values].copy()

    if drop_mt_genes:
        # Remove mitochondrial genes
        gene_names = adata_qc.var_names.values.astype(str)
        mt_gene_mask = ~(np.char.startswith(gene_names, "MT-") | np.char.startswith(gene_names, "mt-"))
        adata_qc = adata_qc[:, mt_gene_mask].copy()
        print("Removed mitochondrial genes. Remaining genes: {len(gene_names)}.")

    return adata_qc
```

```python
def res_search_fixed_clus_leiden(adata, n_clusters, increment=0.01, random_seed=2023):

    for res in np.arange(0.2, 2, increment):
        sc.tl.leiden(adata, random_state=random_seed, resolution=res)
        if len(adata.obs['leiden'].unique()) > n_clusters:
            break
    return res-increment


def leiden(adata, n_clusters, use_rep='SEDR', key_added='SEDR', random_seed=2023):

    sc.pp.neighbors(adata, use_rep=use_rep)
    res = res_search_fixed_clus_leiden(adata, n_clusters, increment=0.01, random_seed=random_seed)
    sc.tl.leiden(adata, random_state=random_seed, resolution=res)

    adata.obs[key_added] = adata.obs['leiden'].astype(int).astype('category')

    return adata
```

```python
from pathlib import Path
from typing import Optional, Tuple, Literal
import os

import numpy as np
import pandas as pd
import scanpy as sc
from scipy.stats import gmean, rankdata
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm, trange
```

```python
def find_neighbors(coor, num_neighbour):
    """
    Find Neighbors of each cell (based on spatial coordinates).
    Args:
        coor: (n_cells, 2), spatial coordinates of cells.
        num_neighbour: number of neighbors to find for each cell.
    Returns:
        spatial_net: with columns "Cell1", "Cell2", and "Distance" representing the neighbors.
    """
    nbrs = NearestNeighbors(n_neighbors=num_neighbour).fit(coor)
    distances, indices = nbrs.kneighbors(coor, return_distance=True)
    cell_indices = np.arange(coor.shape[0])
    cell1 = np.repeat(cell_indices, indices.shape[1])
    cell2 = indices.flatten()
    distance = distances.flatten()
    spatial_net = pd.DataFrame({"Cell1": cell1, "Cell2": cell2, "Distance": distance})
    return spatial_net


def build_spatial_net(adata, num_neighbour):
    """
    Build spatial neighbourhood matrix for each spot (cell) based on the spatial coordinates.
    Args:
        adata: anndata with obsm["spatial"] of shape (n_cells, d).
        annotation: Optional
        num_neighbour: number of neighbors to find per cell.
    Returns:
        Dict: Mapping from cell index to an array of neighbor indices for that cell.
    """


    coor = adata.obsm["spatial"]
    spatial_net = find_neighbors(coor, num_neighbour)

    return spatial_net.groupby("Cell1")["Cell2"].apply(np.array).to_dict()
```

```python
def find_neighbors_regional(cell_pos, spatial_net_dict, coor_latent, num_neighbour):
    """
    Select top-k most similar neighbors among a cell's spatial neighbors based on cosine similarity.
    Args:
        cell_pos: index of the query cell.
        spatial_net_dict: Dict mapping from cell index to its spatial neighbor indices.
        coor_latent: latent embedding (e.g., adata.obsm["SEDR"]) used for cosine similarity.
        config: LatentToGeneConfig object with parameters like num_neighbour and annotation.
    Returns:
       cell_select_pos: selected neighbor indices, sorted by descending similarity.
    """

    cell_use_pos = spatial_net_dict.get(cell_pos, [])
    if len(cell_use_pos) == 0:
        return []

    cell_latent = coor_latent[cell_pos, :].reshape(1, -1)
    neighbors_latent = coor_latent[cell_use_pos, :]
    similarity = cosine_similarity(cell_latent, neighbors_latent).reshape(-1)


    if len(similarity) == 0:
        return []

    indices = np.argsort(-similarity)  # descending order
    top_indices = indices[:num_neighbour]
    cell_select_pos = cell_use_pos[top_indices]
    return cell_select_pos
```

```python
def compute_regional_ranklist(
    cell_pos,
    spatial_net_dict,
    coor_latent,
    ranks,
    frac_whole,
    adata_X_bool,
):
    """
    Compute a neighbor-aggregated rank list for one cell).
    Args:
        cell_pos: index of the query cell.
        spatial_net_dict: Dict mapping from cell index to its spatial neighbor indices.
        coor_latent: latent embedding (e.g., adata.obsm["SEDR"]) used for cosine similarity.
        ranks: (n_cells, n_genes), per-cell per-gene ranks.
    Returns:
        rank_list: a rank list aggregated from neighbors of length n_genes
    """

    cell_select_pos = find_neighbors_regional(
        cell_pos, spatial_net_dict, coor_latent, num_neighbour=21
    )
    if len(cell_select_pos) == 0:
        # Fallback: return this cell's own rank vector as a valid rank list
        return rankdata(ranks[cell_pos, :], method="dense").astype(np.float32, copy=False)


    ranks_tg = ranks[cell_select_pos, :]                    # (k, n_genes)
    agg = gmean(ranks_tg, axis=0)                           # (n_genes,)


    # Ratio of expression fractions
    frac_focal = adata_X_bool[cell_select_pos, :].sum(axis=0).A1 / len(cell_select_pos)
    frac_region = frac_focal / frac_whole
    mask = frac_region <= 1
    agg[mask] = 0

    rank_list = rankdata(agg, method="dense").astype(np.float32, copy=False)
    return rank_list
```

```python
# Initialize
gM = None
frac_whole = None
ranks = np.zeros((n_cells, n_genes), dtype=np.float16)
```

```python
for i in tqdm(range(n_cells), desc="Computing ranks per cell"):
    row = adata_X[i, :].toarray().ravel()
    ranks[i, :] = rankdata(row, method="dense")
```

```python
frac_whole = frac_whole + 1e-12  # Avoid division by zero
```

```python
rank_list_mat = np.zeros((n_cells, n_genes), dtype=np.float32)
```

```python
def _ranklist_wrapper(cell_pos: int) -> np.ndarray:
    return compute_regional_ranklist(
        cell_pos,
        spatial_net_dict,
        coor_latent,
        ranks,
        frac_whole,
        adata_X_bool,
    )
```

```python
for cell_pos in trange(n_cells, desc="Calculating regional rank lists"):
    rank_list_mat[cell_pos, :] = _ranklist_wrapper(cell_pos)
```
